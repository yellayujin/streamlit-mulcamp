```python
import requests
# print(requests.__version__)
req = requests.get("http://google.com")
# req.raise_for_status()
print(req.status_code)   # 응답 코드 확인(잘 불러와졌는지, 200이면 정상)
```

```python
# if문 사용
if req.status_code == requests.codes.ok:    # requests.codes.ok = 200
    print('스크래핑 진행')

# 메서드 사용
req.raise_for_status()   # 200이면 문제 없고, 아니라면 에러를 내보냄
```

```python
# 문제가 없으면 진행, 아니라면 중단하는 코드(요약)
req = requests.get("http://google.com")
req.raise_for_status()
```

```python
# user agent 확인
# https://www.whatismybrowser.com/detect/what-is-my-user-agent/
from bs4 import BeautifulSoup
url = "https://chat.openai.com/c/207aa194-e89a-4a35-8f38-bec9ab0c9a72"
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}
req = requests.get(url, headers = headers)    # 이제 됨
req.raise_for_status()   # 문제가 없으면 진행, 아니라면 중단하는 코드

soup = BeautifulSoup(req.text, 'html.parser')
soup
```
